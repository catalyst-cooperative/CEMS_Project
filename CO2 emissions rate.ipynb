```python
%%sql 
select d.name, d.unitid, d.year, maybe_NG, maybe_coal, CAST(maybe_NG as float) / count(*) as percent_NG, CAST(maybe_coal as float) / count(*) as percent_coal
from data as d, 
(select name, unitid, year, count(*) as maybe_NG from  data where CO2_MASS / heat_input > 0.05 and CO2_MASS / heat_input < 0.07
 group by name, unitid, year) as n,
(select name, unitid, year, count(*) as maybe_coal from  data where CO2_MASS / heat_input > 0.09 and CO2_MASS / heat_input < 0.11
 group by name, unitid, year) as c
WHERE d.name = n.name and d.unitid = n.unitid and d.year = n.year and 
d.name = c.name and d.unitid = c.unitid and d.year = c.year
and d.CO2_MASS > 0
group by d.name, d.unitid, d.year 
```

```python
%%sql
create table n_or_c as 
select d.name, d.unitid, d.op_date, maybe_NG, maybe_coal, CAST(maybe_NG as float) / count(*) as percent_NG, CAST(maybe_coal as float) / count(*) as percent_coal
from data as d, 
(select name, unitid, op_date, count(*) as maybe_NG from  data where CO2_MASS / heat_input > 0.05 and CO2_MASS / heat_input < 0.07
 group by name, unitid, op_date) as n,
(select name, unitid, op_date, count(*) as maybe_coal from  data where CO2_MASS / heat_input > 0.09 and CO2_MASS / heat_input < 0.11
 group by name, unitid, op_date) as c
WHERE d.name = n.name and d.unitid = n.unitid and d.op_date = n.op_date and 
d.name = c.name and d.unitid = c.unitid and d.op_date = c.op_date
and d.CO2_MASS > 0
group by d.name, d.unitid, d.op_date 
```

```python

%sql select * from n_or_c where name = 'Apache Station' and  substr(op_date, 1, 2) is '02' and substr(op_date, -4) is '2016' 
```
