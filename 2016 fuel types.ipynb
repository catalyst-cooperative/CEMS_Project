```python
%%sql 
sqlite:///smaller.db
select substr(op_date, -4) as year, state, fuel_actual, count(distinct(orispl_code)) as num_plants, SUM(heat_input), SUM(CO2_MASS)
FROM (select *,  
    CASE WHEN (fuel_type is 'Natural Gas' or fuel_type is 'Pipeline Natural Gas' or fuel_type is 'Natural Gas, Pipeline Natural Gas') 
    THEN ('NG') ELSE (fuel_type) END as fuel_actual
    FROM data)
group by substr(op_date, -4), state, fuel_actual
```
